```python
import pandas as pd
from sqlalchemy import create_engine

```

```python
super_bowl_file = "Resources/Super_Bowl.csv"
super_bowl_df = pd.read_csv(super_bowl_file)
super_bowl_df.head()
```

```python
# Clean the data by dropping duplicates and setting the index
# super_bowl_df.set_index("SB", inplace=True)
super_bowl_df.head()
```

```python
cleaned_sb_df = super_bowl_df.drop(["Date", "QB  Winner", "Coach Winner", "Coach Loser", "Referee", "Umpire", "Head Linesman", "Line Judge", "Field Judge", "Back Judge", "Side Judge", "Point Difference", "QB Loser"], axis =1)
cleaned_sb_df = cleaned_sb_df[["SB", "Attendance", "Winner", "Winning Pts", "Loser", "Losing Pts", "MVP", "Stadium", "City", "State"]]
cleaned_sb_df = cleaned_sb_df.rename(columns={"Attendance":"attendance", "Winner":"winner", "Winning Pts":"winning_pts", "Loser":"loser", "Losing Pts":"loser_pts", "MVP":"mvp", "Stadium":"stadium", "City":"city", "State":"state"})
cleaned_sb_df.reset_index(drop=True)
cleaned_sb_df.head()
```

```python
connection_string = "postgres:chicago1023@localhost:5432/project2"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
cleaned_sb_df.to_sql(name='superbowl', con=engine, if_exists='append', index=True)
```
